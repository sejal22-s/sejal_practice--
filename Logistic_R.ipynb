# data processing and feature eng

```python
data.isnull().sum()
```

```python
data.loc[data.Pregnancies==0]
```

```python
data.Pregnancies.replace(0,np.median(data.Pregnancies),inplace=True)
```

```python
data.loc[data.Pregnancies==0]
```

```python
data.Glucose.replace(0,np.median(data.Pregnancies),inplace=True)
data.BloodPressure.replace(0,np.median(data.Pregnancies),inplace=True)
data.SkinThickness.replace(0,np.median(data.Pregnancies),inplace=True)
data.Insulin.replace(0,np.median(data.Pregnancies),inplace=True)
data.BMI.replace(0,np.median(data.Pregnancies),inplace=True)
data.DiabetesPedigreeFunction.replace(0,np.median(data.Pregnancies),inplace=True)
data.Age.replace(0,np.median(data.Pregnancies),inplace=True)
```

```python
data.loc[data.Pregnancies==0]
data.loc[data.Glucose==0]
data.loc[data.SkinThickness==0]
data.loc[data.Insulin==0]
data.loc[data.BMI==0]
data.loc[data.DiabetesPedigreeFunction==0] 
data.loc[data.Age==0]
```

```python
plt.figure(figsize=(20,25),facecolor='white')
plotnumber=1
for columns in data:
    if plotnumber <= 16:
        ax = plt.subplot(3,3,plotnumber)
        sns.boxplot(data[columns])
        plt.xlabel(columns,fontsize=20)
        plt.ylabel('Count',fontsize=20)
    plotnumber +=1
plt.tight_layout()
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
dl=['Pregnancies','Outcome']
data1=sc.fit_transform(data.drop(dl,axis=1))
```

```python
con_data=data[['Pregnancies','Outcome']]
```

```python
data.columns
```

```python
type(data1)
```

```python
# type(data1)
data2=pd.DataFrame(data1,columns=['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age'])
```

```python
final_df=pd.concat([data2,con_data],axis=1)
```

```python
final_df
```

# Feature Selection

```python
sns.heatmap(data2.corr(),annot=True)
```

# Model Creation

```python
X=final_df.iloc[:,:-1]
Y=final_df.Outcome
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=43)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train,Y_train)
```

```python
y_predict=LR.predict(X_test)
```

```python
y_predict
```

```python
y_predict_prob=LR.predict(X_test)
```
