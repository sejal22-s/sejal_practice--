# Data Processing

```python
#checking the missing values
data.isnull().sum()
```

# Conversion of the categorical variables

# 1. Attrition

```python
data.Attrition.unique()
```

```python
data.Attrition=data.Attrition.map({'Yes':1,'No':2})
```

```python
data
```

# 2.BusinessTravel

```python
data.BusinessTravel.unique()
```

```python
data.BusinessTravel = data.BusinessTravel.map({"Travel_Rarely":0,"Travel_Frequently":1,"Non-Travel":2})
```

```python
data
```

# 3. Department

```python
data.Department.unique()
```

```python
data.Department = data.Department.map({'Research & Development':0,'Sales':1,'Human Resources':2})
```

```python
data
```

# 4.EducationField

```python
data.EducationField.unique()#unique labels
```

```python
data.EducationField=data.EducationField.map({'Life Sciences':5,'Medical':4,'Marketing':3,
                                             'Technical Degree':2,'Other':1,'Human Resources':0 })
   
```

```python
data.head()
```

# 5.Gender

```python
data.Gender.unique()
```

```python
data.Gender=pd.get_dummies(data.Gender,drop_first=True).astype(int)
```

```python
data.Gender
```

# JobRole

```python
data.JobRole.value_counts()
```

```python
data.JobRole=data.JobRole.map({'Laboratory Technician':8,'Sales Executive':7,
                               'Research Scientist':6,'Sales Representative':5,
                              'Human Resources':4,'Manufacturing Director':3,
                               'Healthcare Representative':2,'Manager':1,
                               'Research Director':0 })
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder#importing label encoder from sklearn 

label = LabelEncoder()#object creation 
data.MaritalStatus=label.fit_transform(data.MaritalStatus)
```

```python
data.MaritalStatus
```

```python
data.OverTime.value_counts()
```
