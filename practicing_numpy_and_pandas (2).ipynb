```python
df1=pd.DataFrame({'A':['A1','A2','A3'],
                 'B':['B1','B2','B3'],
                 'C':['C1','C2','C3'],
                 'D':['D1','D2','D3']})
```

```python
df1
```

```python
df2=pd.DataFrame({'A':['A4','A5','A6'],
                 'B':['B4','B5','B6'],
                 'C':['C4','C5','C6'],
                 'D':['D4','D5','D6']})
df2
```

```python
df3=pd.concat([df1,df2],axis=0,ignore_index=True)
df3
```

```python
df3=pd.concat([df1,df2],axis=1,ignore_index=True)
df3
```

```python
df3=pd.concat([df1,df2],axis=1,ignore_index=False)
df3
```
